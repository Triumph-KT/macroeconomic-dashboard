```python
import pandas as pd
import matplotlib.pyplot as plt

cpi = pd.read_csv('../../data/raw/fred_cpiaucns.csv')
gdp = pd.read_csv('../../data/raw/fred_gdp.csv')
unrate = pd.read_csv('../../data/raw/fred_unrate.csv')
cli = pd.read_csv('../../data/raw/fred_usslind.csv')

for df in [cpi, gdp, unrate, cli]:
    df['date'] = pd.to_datetime(df['date'])
    df['value'] = pd.to_numeric(df['value'], errors='coerce')
```

```python
plt.figure(figsize=(10,5))
plt.plot(cpi['date'], cpi['value'])
plt.title("CPI (Consumer Price Index)")
plt.xlabel("Date")
plt.ylabel("Index Level")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(gdp['date'], gdp['value'])
plt.title("GDP (Gross Domestic Product)")
plt.xlabel("Date")
plt.ylabel("Billions of Dollars (or Index)")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(unrate['date'], unrate['value'])
plt.title("Unemployment Rate (%)")
plt.xlabel("Date")
plt.ylabel("Percent")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(cli['date'], cli['value'])
plt.title("Leading Index for the U.S. -- USSLIND -- (CLI Alternative)")
plt.xlabel("Date")
plt.ylabel("Index Level")
plt.grid(True)
plt.show()
```
